Aim: Develop a model that can classify iris flowers into different species based on their sepal and petal measurements. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans as kk
import seaborn as sns
```

```python
df = pd.read_csv("IRIS.csv")
```

```python
df.tail(10)
```

```python
df['species'], categories = pd.factorize(df['species'])
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
x = fig.add_subplot(111, projection='3d')
x.scatter(df.petal_length, df.petal_width, df.species)
x.set_xlabel('Petal Length (in cm)')
x.set_ylabel('Petal Width (in cm)')
x.set_zlabel('species')
plt.title('3D Scatter Plot')
plt.show()
```

```python
sns.scatterplot(data = df,x="sepal_length", y="sepal_width", hue= "species")
```

```python
sns.scatterplot(data = df, x='petal_length', y= 'petal_width', hue='species')
```

```python
rng = range(1,10)
sse = []

for i in rng:
    km = kk(n_clusters=i)
    km.fit(df[['petal_length', 'petal_width']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel("k_range")
plt.ylabel("Sum of Squared errors")
plt.plot(rng, sse)
```

The above graph shows that the elbow is at the 3rd number

```python
km = kk(n_clusters=3, random_state= 0,)
y_predicted = km.fit_predict(df[['petal_length', 'petal_width']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head(150)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(df.species, df.cluster)
cm
```

```python
true_labels = df.species
predicted_labels = df.cluster
cm = confusion_matrix(true_labels, predicted_labels)
class_labels = ["Setosa", "Versicolor", "Virginica"]
#Plot Confusion Matrix
plt.imshow(cm, interpolation = 'nearest', cmap = plt.cm.Blues)
plt.title("Confusion Matrix")
plt.colorbar()
tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks, class_labels)
plt.yticks(tick_marks, class_labels)

for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        plt.text(j, i, str(cm[i][j]), ha = 'center', va = 'center', color = 'white')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
